12. **Analysis and Data Visulization**

```python
first_confirmed_cases = covid_bangladesh_df.date.loc[covid_bangladesh_df['new_cases'].ne(0).idxmax()]
print("THE First Confirmed Case in Bangladesh was reported on {}.". format(str(first_confirmed_cases)))
```

13. How Many Cases were reported by the time the lock down was in implemented(March 10, 2020)


```python
#Date when the first lockdown was implemented
```

```python
# Date when the locked down was implemented

locked_down_date = '2020-03-16'
total_cases_lock_down_date = covid_bangladesh_df.total_cases.loc[covid_bangladesh_df.loc[covid_bangladesh_df.date == locked_down_date].index.values.astype(int)[0]]
print("The total number of confirmed cases in Bangladesh by {} was {}.". format(locked_down_date, int(total_cases_lock_down_date)))
```

14. How many deaths were reported by the time locked down was implemented ?

```python

total_deaths_locked_down_date = covid_bangladesh_df.total_deaths.loc[covid_bangladesh_df.loc[covid_bangladesh_df.date == locked_down_date].index.values.astype(int)[0]]
print("The total number of deaths by {} was {}.".format(locked_down_date, int(total_deaths_locked_down_date)))




```

15. When was the first death related to Covid-19 reported

16. How did the number daily new cases and new deaths evolded in Bangladesh ?


```python
plt.plot(covid_bangladesh_df.new_cases, 'b-', linewidth=1, markersize=5, markeredgewidth=2, markeredgecolor='blue')

plt.plot(covid_bangladesh_df.new_deaths, 'r--', linewidth=1, markersize=5,)
plt.title(' Covid19 Bangladesh Daily positive cases vd Daily Deaths')
plt.legend(['Daily cases', 'Daily deaths'])
plt.xlabel('Days');plt.ylabel('count');
```

From the above figure, it looks like that at the early stages of the locked fown the new deaths reported considerately low, compared to the number of cases, but in a few days later a big spike in cases arise. also it seemes that today Bangladesh is getting as much new daily cases as the country reporting during the lockdown, but the number of deaths are very close to 0.
